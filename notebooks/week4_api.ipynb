# Week 4: API Development, Model Deployment & Final Evaluation

## Objective

The goal of Week 4 is to deploy the trained machine learning models as a real-time API.

This transforms the project into a production-ready AI system capable of:

- Accepting user complaints
- Predicting department automatically
- Performing sentiment analysis
- Assigning priority scores

---

## System Workflow

1. User submits complaint text
2. Text is cleaned and processed
3. TF-IDF converts text to numerical features
4. ML model predicts department
5. BERT model analyzes sentiment
6. Priority score is assigned
7. API returns structured JSON response

---

## Technology Used

- FastAPI → API development
- Joblib → Model loading
- Scikit-learn → ML model
- Transformers (BERT) → Sentiment analysis

---

## Expected Outcome

A fully functional REST API that performs real-time grievance classification and prioritization.
