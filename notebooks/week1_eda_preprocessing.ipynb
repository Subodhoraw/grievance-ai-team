## Step 1: Loading the Dataset
In this step, the raw grievance dataset in JSON format is loaded into a pandas DataFrame.
Since the dataset contains structured records, we convert it into a tabular format for easier processing and analysis.

````python
import pandas as pd
import json

file_path = '../data/raw/no_pii_grievance.json'

with open(file_path, 'r', encoding='utf-8') as f:
    data = json.load(f)

df = pd.DataFrame(data)

df.head()
````

##  Step 2: Understanding the Data Structure

Before cleaning the data, it is important to understand its structure.

We inspect:
- Column names
- Sample records
- Data types

This helps identify which fields are useful for our analysis and modeling.

````python
df.columns
````

##  Step 3: Selecting Relevant Features
The dataset contains multiple fields, but not all are required for this project.
We focus on:
- Complaint text (input feature)
- Department (target variable)
This simplifies the dataset and makes it suitable for machine learning.

````python
df.head(2).T
````

## Step 4: Renaming Columns
To improve readability and maintain consistency across the project, column names are simplified.
- `subject_content_text` → `complaint`
- `org_code` → `department`
This makes the dataset easier to work with and understand.

````python
df_clean = df[['subject_content_text', 'org_code']]
df_clean.head()
````

## Step 5: Handling Missing Values
Real-world datasets often contain missing or incomplete records.
Rows with missing values in important columns are removed to ensure:
- Data quality
- Reliable model performance

````python
df_clean.rename(columns={
    'subject_content_text': 'complaint',
    'org_code': 'department'
}, inplace=True)
df_clean.head()
````

## Step 6: Text Cleaning
The complaint text is cleaned to remove noise and standardize the data.
Cleaning steps include:
- Converting text to lowercase
- Removing special characters and symbols
- Eliminating unnecessary formatting
This prepares the text for further NLP processing.

````python
df_clean = df_clean.dropna()
df_clean.isnull().sum()
````

## Step 7: Saving Cleaned Data
After preprocessing, the cleaned dataset is saved for future use.
This ensures:
- Separation of raw and processed data
- Reusability without repeating preprocessing steps

````python
import re
def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'\r\n', ' ', text)
    text = re.sub(r'>>', ' ', text)
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    return text
df_clean['complaint'] = df_clean['complaint'].apply(clean_text)
df_clean.head()
````

## Step 8: Basic Data Exploration
To gain initial insights, we analyze the distribution of complaints across departments.
This helps in:
- Understanding data imbalance
- Identifying dominant categories
- Preparing for model training

````python
df_clean.to_csv('../data/processed/cleaned_data.csv', index=False)
````

````python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

# Clean text
text = " ".join(df_clean['complaint'].dropna())

# Create WordCloud
wordcloud = WordCloud(
    width=1000,
    height=500,
    stopwords=STOPWORDS,
    background_color='white',
    colormap='viridis'
).generate(text)

# Display
plt.figure(figsize=(10,5))
plt.imshow(wordcloud)
plt.axis('off')
plt.title("Customer Complaints WordCloud")
plt.show()
````

## N-Gram Analysis
N-grams help identify frequently occurring word combinations in complaints.
This provides deeper insights into common issue patterns such as "water supply", "road damage", etc.

````python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(ngram_range=(2,2))  # bigrams
X = vectorizer.fit_transform(df_clean['complaint'])
sum_words = X.sum(axis=0)
words_freq = [(word, sum_words[0, idx]) for word, idx in vectorizer.vocabulary_.items()]
words_freq = sorted(words_freq, key=lambda x: x[1], reverse=True)
words_freq[:10]
````
